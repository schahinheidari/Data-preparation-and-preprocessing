```python
dfNew = df['Win-Loss'].str.split('-', expand=True)
```

```python
dfNew
```

```python
df['Win'] = dfNew[0]
df['Loss'] = dfNew[1]
```

```python
df.head()
```

```python
df.drop(columns=['Win-Loss'], inplace=True)
df.head()
```

```python
team = df['Team']
team.head()
```

```python
# https://pandas.pydata.org/docs/reference/api/pandas.to_numeric.html
# errors='coerce' اگر جایی نتوانستی تبدیل کنی بجاش نال قرار بده یا 
# NaN
df[['Win']] = df[['Win']].apply(pd.to_numeric, errors='coerce')
```

```python
df.dtypes
```

```python
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.corr.html
dfCorr = df.corr()
dfCorr
```

```python
dfCorr['Win']
```

```python
df[['Loss']] = df[['Loss']].apply(pd.to_numeric, errors='coerce')
```

```python
dfCorr = df.corr()
```

```python
dfCorr['Loss']
```

```python
# وقتی دیتاهای ما عددی هستند  و رتبه ای نیستند از متد پیرسون استفاده بکنیم بهتر است
# pearson
dfCorr['Loss']['Off Plays']
```

```python
# وقتی دیتاهای ما رتبه ای هستند بهتر است از متدهای زیر استفاده کنیم
# spearman ----- Kendall
dfCorr = df.corr(method='spearman')
```

```python
dfCorr['Off Plays']
```

```python
# اگر دیتامون نویز داشته باشه و همینطور رنکی هم هست از کندال استفاده کنیم
dfCorr = df.corr(method='kendall')
```

```python
dfCorr['Loss']['Off Rank']
```

```python
dfCorr['Loss']['Off Plays']
```

```python
dfCorrK = df.corr(method='kendall')
```

```python
# وقتی در اینجا مثلا بخواهیم کورولیشن برنده رو با مثلا آف پلی مقایسه کنیم بایستی از متد پیرسون استفاده کنیم 
# ولی وقتی مثلا برای مقایسه کورولیشن برنده با آف رنک بخواهیم از متدی استفاده کنیم بهترین متد میشه گفت کندال است
dfCorrK['Win']
```
